```python
youtubeData = pd.read_csv(io.BytesIO(uploaded['most_subscribed_youtube_channels.csv']))
display(youtubeData)
```

```python
#remove the commas from the integers
youtubeData = youtubeData.replace(',', '', regex=True)
display(youtubeData)
```

```python
#Histogram
import plotly.express as px

fig = px.histogram(youtubeData, x='subscribers', title = 'Subscriber Count')
fig.show()
```

```python
#pie chart
fig = px.pie(youtubeData, values= 'subscribers', names= 'category', title= 'Youtube Categories')
fig.show()
```

```python
#Box plot

fig= px.box(youtubeData, y='started', title= 'Years Started')
fig.show()
```
